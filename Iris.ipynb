# Loading Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
```

# load Iris Dataset

```python
iris = datasets.load_iris()
#Convert to DataFrame
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "target"]
df = pd.DataFrame(data=np.c_[iris.data, iris.target], columns=columns)
mapping = {0:'setosa', 1:'versicolor', 2:'virginica'}
df["target"]=df["target"].map(mapping)
df
```

```python
df.describe()
```

# plot

```python
columns = columns[:4]
x_data = range(0, df.shape[0])
for column in columns:
    plt.plot(x_data, df[column], label=column)
plt.title('Iris', size = 15)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Scatter

```python
plt.figure(figsize = (4,2))
plt.scatter(df["sepal_length"],df["sepal_width"])
plt.show()
```

```python
colors = mapping = {'setosa':"b", 'versicolor':"r", 'virginica':"g"}
plt.figure(figsize = (4,2))
for i in range(150):
    plt.scatter(df['sepal_length'][i], df['sepal_width'][i], c = colors[df['target'][i]])

plt.title('Iris Dataset')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

# Histogran

```python
i = 1
for col in columns:
    plt.subplot(2, 2, i)
    plt.hist(df[col], bins=12)
    plt.title(col)
    i = i + 1
```

# 3D

```python
ax = plt.axes(projection="3d")
x_data = np.arange(0, 20, 0.2)
x_data = np.arange(0, 20, 0.2)
for i in range(150):
    plt.scatter(df['petal_width'][i], df['sepal_width'][i],df["petal_length"][i], c = colors[df['target'][i]])
#ax.scatter(df["petal_width"], df["petal_length"], df["sepal_length"])
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_length')
plt.show()
```
