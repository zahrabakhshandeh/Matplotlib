```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plot

```python
x = np.array([10, 5, 17])
y = np.array([1, 15, 7])
plt.figure(figsize = (5,3)) #2
#plt.plot(x,y, label = "x")
#plt.plot(x, y, label = "2x", color = "red", linewidth = 2, linestyle = "--",
#        marker="o", markersize=10,markeredgecolor = "blue") #6
plt.plot(x, y, 'ro--', label = "xy", ms = 5) #7
plt.xlabel("X", c = "r", size = 20) #3 #4
plt.ylabel("Y") #3
plt.title("first plot", c = "b", size = 20) #3
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5, axis = "x") #9
plt.legend() #5
plt.savefig("img.png")
plt.show()
```

```python
x = np.linspace(-20.0, 20.0, num=20)
plt.figure(figsize = (5,3))
plt.plot(x, x , 'b--', label = "2x")
plt.plot(x, x * 4, 'y:', label = "2x")
plt.plot(x, x ** 2, 'ro--', label = "x^2", ms=1)
plt.title("First plot!", size = 20, color = "red")
plt.xlabel('X Axis', size = 20, color = "red")
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

# Scatter

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.figure(figsize = (5,3))
plt.scatter(x, y, s = sizes)
plt.title("Scatter!", size = 20, color = "blue")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = np.array([1, 3, 6, 10, 14])
y = np.array([70, 60, 50, 40, 30])
plt.figure(figsize = (5,3))
plt.scatter(x, y, color = 'hotpink')

x = np.array([30, 4, 17, 3])
y = np.array([4, 15, 0, 70])
plt.scatter(x, y, color = '#88c999')
plt.show()
```

```python
x = np.array([1, 3, 6, 10, 14])
y = np.array([20, 30, 50, 10, 4])
colors = ["blue", "red", "black", "yellow", "pink"]
plt.figure(figsize = (5,3))
plt.scatter(x, y, color = colors)
plt.show()
```

# Pie Charts

```python
x = np.array([4,6,100])
y = np.array(["c1", "c2", "c3"])
myexplode = [0, 0.2, 0]
plt.figure(figsize = (5,3))
plt.pie(x,labels = y, autopct='%1.3f%%', explode = myexplode)
plt.show()
```

# Bars

```python
x = np.array(["C1", "C2", "C3"])
y = np.array([4, 6, 100])
plt.figure(figsize = (5,3))
plt.bar(x, y, width = 0.1)
plt.show()
```

```python
x = np.array(["C1", "C2", "C3"])
y = np.array([4, 6, 100])
plt.figure(figsize = (5,3))
plt.barh(x, y, height = 0.1, color = "green")
plt.show()
```

```python
age = np.array([20, 23, 14, 20, 20, 15])
plt.figure(figsize = (5,3))
plt.hist(age)
plt.show() 
```

```python
x = np.random.normal(1, 100, size=1000)
plt.figure(figsize = (5,3))
plt.hist(x)
plt.show() 
```

# boxplot

```python
plt.boxplot(x)
plt.show()
```

# subplot

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

plt.subplot(2, 3, 2)
plt.pie(y)

plt.subplot(2, 3, 3)
plt.pie(x)

plt.subplot(2, 3, 4)
plt.hist(y)

plt.subplot(2, 3, 5)
plt.scatter(x,y)

plt.subplot(2, 3, 6)
plt.hist(x)
plt.show()
```

# 3D

```python
ax = plt.axes(projection="3d")
x_data = np.arange(0, 20, 0.2)
y_data = np.arange(20, 40, 0.2)
z_data = np.sin(x_data) * np.cos(y_data)
ax.plot(x_data, x_data,z_data)
#ax.scatter(df["petal_width"], df["petal_length"], df["sepal_length"])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.savefig("foo.png")
plt.show()
```
